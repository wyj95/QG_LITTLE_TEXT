```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
from QG_week2_Data_class import Data


def drank(data_frame, title):
    Xmat = data_frame.values
    feature_list = data_frame.columns.values
    col = np.shape(Xmat)[1]
    col -= 1
    i = random.randint(0, col - 1)
    plt.title(title)
    plt.xlabel(feature_list[i])
    plt.ylabel("Price")
    plt.scatter(Xmat[:, i], Xmat[:, -1], marker='*')
    plt.show()

boston_load = load_boston()
data_frame = pd.DataFrame(boston_load.data)
data_frame.columns = boston_load.feature_names
data_frame["PRICE"] = boston_load.target
drank(data_frame, "Boston")
boston = Data(data_frame)
print("在未标准化之前：")
print("其准确的系数矩阵为：")
print(boston.true_line())
print("其损失函数值为：")
print(boston.get_f())
print("其梯度系数矩阵为：")
print(boston.grad_line())
print("其损失函数值为：")
print(boston.get_f())
print("不标准化,迭代次数多,效果还差")
print(" ")

print("max-min标准化后：")
boston.max_min_change()
print("其准确系数矩阵为：")
print(boston.true_line())
print("其损失函数值为：")
print(boston.get_f())
print("其梯度系数矩阵为：")
print(boston.grad_line(n=0.00001, max_t=100000, min_change=0.001))
print("其梯度损失函数值为：")
print(boston.get_f())
print(" ")

boston.normol_change()
print("在正态标准化后：")
print("其准确的系数矩阵为：")
print(boston.true_line())
print("其损失函数值为：")
print(boston.get_f())
print("其梯度系数矩阵为：")
print(boston.grad_line(n=0.0005, max_t=1000, min_change=0.001))
print("其损失函数值为：")
print(boston.get_f())
```
